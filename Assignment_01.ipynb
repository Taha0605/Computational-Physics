## Question 2 ##
What happens if we use the Newton-Raphson Method to obtain the roots of the equation, f(x) = x
3 − x − 3
with initial guess, x0 = 0.

```python
def FofX(x):
    fox = math.pow(x,3)-x-3
    return fox

def FdOfX(x):
    fdox = 3*x*x -1
    return fdox

def newX(x):
    return(x - (FofX(x)/FdOfX(x)))

print("Function : x^3 - x -3")
x = float(input("Enter the initial value of x: "))
counter = 0
while (counter<150):
    x = newX(x)
    # print(x)
    counter+= 1
print("Root found at ",x) 
```

where a = 10, c = 8/3 are some constants. b denotes the temperature difference between the top and bottom
surfaces of the fluid. Solve the equations with RK-4 method with a = 10, c = 8/3 and b = 5, 10, 25. Plot the
following:<br>
(a) z as a function of time for b = 5, 10, 25 for x = 1, y = z = 0. You can move from t = 0s to t = 50s. Is
there any stricking difference at b = 25?<br>
(b) The trajectory of Lorenz model (for b = 25) in x − z plane with initial condition as x = 1, y = z = 0.<br>
(c) The trajectory in yz plane when x = 0 with b = 25.<br>
(d) The trajectory in xz plane when y = 0 with b = 25<br>

```python
a = 10
b_range = [5, 10, 25]
c = 8/3
```

```python
## Function definitions ##
def fx(x, y, z): ## dx/dt
    v = a*(y-x)
    return v
```

```python
def fy(x, y, z): ## dy/dt
    v = -1*x*z + b_range[2]*x - y
    return v
```

```python
def fz(x, y, z): ## dz/dt
    v = x*y - c*z
    return v
```

```python
## Implementation of Runge-Kutta Method ##
def RK(x, y, z, t):
  x_points = []
  y_points = []
  z_points = []

  while t < 50:  

    kx1 = timestep*fx(x, y, z)
    ky1 = timestep*fy(x, y, z)
    kz1 = timestep*fz(x, y, z)

    kx2 = timestep*fx(x+kx1*0.5, y+ky1*0.5, z+kz1*0.5)
    ky2 = timestep*fy(x+kx1*0.5, y+ky1*0.5, z+kz1*0.5)
    kz2 = timestep*fz(x+kx1*0.5, y+ky1*0.5, z+kz1*0.5)

    kx3 = timestep*fx(x+kx2*0.5, y+ky2*0.5, z+kz2*0.5)
    ky3 = timestep*fy(x+kx2*0.5, y+ky2*0.5, z+kz2*0.5)
    kz3 = timestep*fz(x+kx2*0.5, y+ky2*0.5, z+kz2*0.5)

    kx4 = timestep*fx(x+kx3, y+ky3, z+kz3)
    ky4 = timestep*fy(x+kx3, y+ky3, z+kz3)
    kz4 = timestep*fz(x+kx3, y+ky3, z+kz3)

    x = x + (kx1 + 2*kx2 + 2*kx3 + kx4)/6
    y = y + (ky1 + 2*ky2 + 2*ky3 + ky4)/6
    z = z + (kz1 + 2*kz2 + 2*kz3 + kz4)/6

    # x = x + timestep*kx1
    # y = y + timestep*ky1
    # z = z + timestep*kz1

    x_points.append(x)
    y_points.append(y)
    z_points.append(z)

    t += timestep
  #plot_points(x_points, y_points, z_points)
  #plot_points_xz(x_points, z_points)
  plot_points_xz(y_points, z_points)
```
